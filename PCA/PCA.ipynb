```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python

iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns =  iris.feature_names)
iris_df
```

```python
X = iris.data            # features
y  = iris.target        # label  - data you are trying to predict or classify

df = pd.DataFrame(X, columns = iris.feature_names)
df["species"] = iris.target_names[y]
df
```

```python
# so our features are whole data except "species" and our target is species
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)    # we standarddized the features

pca = PCA(n_components = 2)
X_pca = pca.fit_transform(scaled_data)

df_pca = pd.DataFrame(X_pca, columns = ['PC1', 'PC2'])
df_pca['species'] = y
df_pca 


```

```python
# so here main this is we have converted all those features into two PCA's which is PCA 1 and PCA 2
# this two can be used for further ML models to get insights in fast speed and correct output.
```
