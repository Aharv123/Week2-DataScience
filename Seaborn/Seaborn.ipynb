```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    "name": ["Atharv", "Viraj", "Abdul", "Joseph"],
    "marks": [78,88,98,69]
    
}
df = pd.DataFrame(data)

sns.barplot(x = df["name"], y = df["marks"])
plt.title("Bar Plot")
plt.show()

# if we observed one thing In matplot we basically define labels for x and y but using seaborn 
# it automatically gives x and y labels 
```

```python
import numpy as np
arr1 = np.array(["A", "B", "C"])
arr2 = np.array([78,54,76])


sns.barplot(x = arr1 ,y = arr2)
plt.title("bar plot")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {

    "name": ["Atharv", "Arjun", "Rampal"],
    "marks": [23,43,34],
    "Gender":["M", "F", "F"]
}

df =  pd.DataFrame(data)


sns.scatterplot(data=df, x="name", y="marks", hue="Gender", style="Gender", size="marks", sizes=(50, 200))
plt.show()
```

```python

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Sales": [200, 250, 300, 350, 400],
    "Region": ["North", "North", "South", "South", "North"]
}

df = pd.DataFrame(data)

sns.lineplot(data=df, x="Month", y="Sales", hue="Region", style="Region", markers=True)
plt.title("Line Plot: Monthly Sales by Region")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 30, 30, 40, 50, 50, 50, 50, 60]

sns.histplot(data, bins=5, kde=True,  color='skyblue')
plt.title("Histogram with KDE")
plt.show()
```

```python
# sns.kdeplot()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 30, 30, 40, 50, 50, 50, 50, 60]

sns.kdeplot(data, fill=True, color='purple')
plt.title("KDE Plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 30, 30, 40, 50, 50, 50, 50, 60]

sns.histplot(data, kde=True, bins=10, color='orange', alpha=0.6)
plt.title("Histogram with KDE Overlay")
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("iris")
data
```

```python
sns.pairplot(data, hue="species", diag_kind="kde", palette="husl")
plt.show()
```
