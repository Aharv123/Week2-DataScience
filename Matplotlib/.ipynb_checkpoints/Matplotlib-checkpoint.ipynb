```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, color='red', linestyle='--', marker='o', label='Sine Wave')
plt.title("Customized Line Plot", fontsize=14)
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.grid(True, linestyle=':', linewidth=0.5)
plt.legend()
plt.savefig("Customized_Line_Plot.jpg")
plt.show()
```

```python
# Annotations
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='Sine Wave')
plt.title("Annotated Plot")
plt.annotate('Peak', xy=(1.5,1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.legend()
plt.savefig("Annotated_plot.png",bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data2 = {
    "Name": ["Rampal",  "Joseph", "Akash"],
    "Age": [56,45,23]
}

df = pd.DataFrame(data2)

plt.bar(x = df["Name"], height = df["Age"], color = "purple", edgecolor = "black")
plt.title("Age Chart")
plt.xlabel("Name")
plt.ylabel("Age")
plt.savefig("AGE_bar.png", dpi = 300, bbox_inches='tight')
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

data = {
    "Product": ["Chargers", "Mouse", "Laptops", "Mouse Pads"],
    "Sales": [50, 20,20 ,10 ]
}

df = pd.DataFrame(data)
plt.pie(df["Sales"], labels = df["Product"],autopct= '%1.1f%%',  startangle = 90)
plt.title("Sales (%)")
plt.savefig('Sales_Chart.jpg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python


# Data
data = {
    "Product": ["Chargers", "Mouse", "Laptops", "Mouse Pads"],
    "Sales (in $)": [100, 700, 500, 178]
}

df = pd.DataFrame(data)

def format_dollars(value):
    absolute = int(value / 100 * sum(df["Sales (in $)"]))  
    return f"${absolute}"  
    
plt.pie(df["Sales (in $)"], labels=df["Product"], autopct=format_dollars, startangle=90)
plt.title("Sales Distribution (in $)")
plt.savefig("Sales_$.png")
plt.show()
```

```python
# To show it in 3D we do like this
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Product": ["Chargers", "Mouse", "Laptops", "Mouse Pads"],
    "Sales (in $)": [100, 700, 500, 178]
}

df = pd.DataFrame(data)

def format_dollars(value):
    absolute = int(value / 100 * sum(df["Sales (in $)"]))  
    return f"${absolute}"  

plt.figure(figsize=(8, 6))
plt.pie(
    df["Sales (in $)"], 
    labels=df["Product"], 
    autopct=format_dollars, 
    startangle=90, 
    explode=(0, 0.1, 0, 0),  # "explode" one slice to highlight
    shadow=True  # Add shadow for 3D effect
)

plt.title("3D Sales Pie Chart")
plt.savefig("3D_Sales_Pie_Chart.jpg")
plt.show()
```
