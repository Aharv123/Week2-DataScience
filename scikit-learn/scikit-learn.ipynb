```python
# Feature Scaling

# Feature scaling is the process of adjusting the values 
# of different features (columns) in your dataset to a common range or scale.
# This ensures no single feature dominates or biases the 
# model because of its larger scale compared to others.

# In short it treats all the features to same level and equally.
```

```python
# Types

# StandardScaler (Z score Normalization)
# MinMaxScaler
# RobustScaler
```

```python
# StandardScaler - Standardizes features by removing the mean 
# and scaling to unit variance.

# X scaled = X−μ / σ
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler  # imp
from sklearn.datasets import load_wine 

wine = load_wine()

df = pd.DataFrame(wine.data, columns = wine.feature_names)

scaler = StandardScaler()

standardize_data = scaler.fit_transform(df)

x = pd.DataFrame(standardize_data, columns = df.columns)
x
```

```python
# MinMaxScaler - Scales features to a specific range, typically [0, 1].

# X scaled = X - X min / X max - X min
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

data = {

    "Age": [23,34,56,89],
    "Salary": [20000,70000,76000,87000],
    "House Size": [900, 1500, 2500, 3500]
}

df = pd.DataFrame(data)

scaler = MinMaxScaler()

y = scaler.fit_transform(df)

z = pd.DataFrame(y, columns = df.columns)
z
```

```python
# RobustScaler

# Scales features based on their median and interquartile range (IQR).

# X scaled = X−Median / IQR

# IQR = Q3 -Q1

# Q1 = value at the 25th percentile of the data
# Q3 = value at the 75th percentile of the data
```

```python
import pandas as pd
from sklearn.preprocessing import RobustScaler
from sklearn.datasets import load_iris


iris = load_iris()

df = pd.DataFrame(iris.data, columns = iris.feature_names)

robust = RobustScaler()

robusted_data = robust.fit_transform(df)

result = pd.DataFrame(robusted_data, columns=df.columns)
result
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('Housing.csv')

# First train_test_split
X = df.drop(columns = ['price'])
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)

print("Feature Train:\n", X_train)
print("\nFeature Test:\n", X_test)
print("\nTarget Train:\n", y_train)
print("\nTarget Test:\n", y_test)

print("============================================================================================================")

# Now do Preprocessing using any Feature Scaling technique

numerical_cols = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking']

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train[numerical_cols])

X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=numerical_cols)

print("\nScaled Feature Train Data:\n", X_train_scaled_df)
```
