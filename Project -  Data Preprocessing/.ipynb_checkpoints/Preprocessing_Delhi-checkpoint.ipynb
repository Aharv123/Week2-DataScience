```python
import pandas as pd

data = pd.read_csv("delhi_aqi.csv")
data
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.feature_selection import mutual_info_regression
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder, OneHotEncoder, PolynomialFeatures

# we will first convert date column to datetime
data["date"] = pd.to_datetime(data['date'])
data["date"]

# Dataset Info
print("Dataset Info:")
print(data.info())
print("\nPreview of Dataset:")
print(data.head())

# Handling missing values
print("\nMissing Values:")
print(data.isnull().sum())

data.fillna(data.mean(), inplace=True)

# Extract date features
data['date'] = pd.to_datetime(data['date'])
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
data['hour'] = data['date'].dt.hour
data.drop('date', axis=1, inplace=True)

# Exploratory Data Analysis (EDA)
print("\nSummary Statistics:")
print(data.describe())

# Distribution plots for numerical columns
numerical_columns = ['co', 'no', 'no2', 'o3', 'so2', 'pm2_5', 'pm10', 'nh3']
for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(data[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# Boxplot for outliers
for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data[col])
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()

# Bar Plot: Average values of pollutants
plt.figure(figsize=(10, 6))
data[numerical_columns].mean().plot(kind='bar', color='skyblue')
plt.title("Average Values of Pollutants")
plt.xlabel("Pollutants")
plt.ylabel("Average Value")
plt.xticks(rotation=45)
plt.show()

# Line Plot: Trends of PM2.5
plt.figure(figsize=(10, 6))
plt.plot(data['pm2_5'], color='blue', label='PM2.5')
plt.title("Trend of PM2.5 Over Records")
plt.xlabel("Index")
plt.ylabel("PM2.5 Levels")
plt.legend()
plt.show()

# Histogram Plot: PM2.5
plt.figure(figsize=(8, 5))
plt.hist(data['pm2_5'], bins=30, color='orange', alpha=0.7, edgecolor='black')
plt.title("Histogram of PM2.5")
plt.xlabel("PM2.5 Levels")
plt.ylabel("Frequency")
plt.show()

# Feature Extraction
X = data[numerical_columns]
y = data['pm2_5']  # Target variable (example)
mutual_info = mutual_info_regression(X, y)
mutual_info_series = pd.Series(mutual_info, index=numerical_columns)
print("\nFeature Importance (Mutual Information):")
print(mutual_info_series.sort_values(ascending=False))

# Correlation Analysis
correlation_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()

# Identify highly correlated features
threshold = 0.8
high_corr_features = correlation_matrix[(correlation_matrix >= threshold) & (correlation_matrix < 1.0)]
print("\nHighly Correlated Features (above 0.8):")
print(high_corr_features)

# Binning
bins = [0, 50, 100, 300, np.inf]
labels = ['Good', 'Moderate', 'Poor', 'Very Poor']
data['pm2_5_category'] = pd.cut(data['pm2_5'], bins=bins, labels=labels)
print("\nPM2.5 Binning Value Counts:")
print(data['pm2_5_category'].value_counts())

# Bar Plot for PM2.5 Categories
plt.figure(figsize=(8, 5))
sns.countplot(data['pm2_5_category'])
plt.title("PM2.5 Category Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()

# Feature Scaling: Standardization
scaler = StandardScaler()
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])
print("\nAfter Standardization:")
print(data[numerical_columns].head())

# Feature Scaling: Normalization
normalizer = MinMaxScaler()
normalized_data = normalizer.fit_transform(data[numerical_columns])
normalized_df = pd.DataFrame(normalized_data, columns=numerical_columns)
print("\nAfter Normalization:")
print(normalized_df.head())

# Encoding: Label Encoding
encoder = LabelEncoder()
data['pm2_5_category_encoded'] = encoder.fit_transform(data['pm2_5_category'])
print("\nLabel Encoded PM2.5 Category:")
print(data[['pm2_5_category', 'pm2_5_category_encoded']].head())

# Encoding: One-Hot Encoding
one_hot = pd.get_dummies(data['pm2_5_category'], prefix='pm2_5_category')
data = pd.concat([data, one_hot], axis=1)
print("\nOne-Hot Encoded PM2.5 Category:")
print(one_hot.head())

# Feature Transformation: Polynomial Features
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(data[numerical_columns])
poly_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(numerical_columns))
print("\nPolynomial Features:")
print(poly_df.head())

# Feature Transformation: Log Transformation
log_transformed = data[numerical_columns].apply(lambda x: np.log1p(x))
print("\nLog Transformed Features:")
print(log_transformed.head())

# Save Preprocessed Data
data.to_csv("delhi_sqi_processed.csv", index=False)
print("\nProcessed data saved to 'delhi_sqi_processed.csv'.")
```
