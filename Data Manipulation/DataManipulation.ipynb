```python
import pandas as pd
```

```python
df = pd.read_csv("student-dataset.csv")
df
```

```python
# using apply() to transform a data.  Note-  it can only be used on DataFrame
```

```python
df["age"] = df["age"].apply(lambda y: y + 4 )
df
```

```python
# map() is used for transforming data.  Note -  it can only be used on Series
```

```python
df["city"] = df["city"].map(lambda y: y.upper())
df
```

```python
df["city"] = df["city"].map(lambda y: y.lower())
df
```

```python
df["language.grade"] = df["language.grade"] + 4
df
```

```python
# 5] Pivot Tables and Grouping
```

```python
# Pivot Tables  ->  It reorganizes data into understandable and proper format
```

```python
data = {
    "Employee": ["Alice", "Bob", "Charlie", "Alice", "Bob"],
    "Department": ["HR", "IT", "Finance", "HR", "IT"],
    "Year": [2022, 2022, 2022, 2023, 2023],
    "Salary": [50000, 60000, 55000, 52000, 61000],
    "Bonus": [5000, 6000, 5500, 5200, 6100],
}
df = pd.DataFrame(data)
df
```

```python
pd.pivot_table(df, values="Salary", index = "Department", columns = "Year", aggfunc = "mean")
```

```python
# groupby()
```

```python
df.groupby("Department")["Salary"].sum()
```

```python
df.groupby("Employee")["Salary"].sum()
```

```python
df.groupby("Year")["Employee"].size()
```

```python
df.groupby("Department")[["Salary","Bonus"]].sum()
```

```python
# Combining Pivot Tables and Groups
```

```python
pivot_table = df.groupby(["Department", "Year"])["Salary"].mean().unstack()
pivot_table
```

```python
# Sorting
# we use   sort_values()   sort_index()
```

```python
data = {
    "Employee": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Department": ["HR", "IT", "Finance", "HR", "IT"],
    "Salary": [52000, 60000, 55000, 50000, 61000],
    "JoiningDate": ["2022-01-15", "2021-05-20", "2020-09-10", "2023-03-01", "2020-11-25"],
    "Sales": [200, 220, 250, 230, 290],
}


df = pd.DataFrame(data)
df
```

```python
df.sort_index(axis = 1, ascending = True)
```

```python
df.sort_index(ascending = False)
```

```python
df.sort_values(by = "Salary", ascending = False)
```

```python
df.sort_values(by = ["Department", "Salary"], ascending = [True, False])
```
